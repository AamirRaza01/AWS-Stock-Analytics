```python
import boto3
```

```python
AWS_ACCESS_KEY = ""
AWS_SECRET_KEY = ""
AWS_REGION = "ap-south-1"
MY_BUCKET_NAME = ""
GLUE_SERVICE_ROLE = ""
S3_PATH = f"s3://{MY_BUCKET_NAME}/"
```

```python

glue = boto3.client(
    'glue',
    region_name = AWS_REGION,
    aws_access_key_id = AWS_ACCESS_KEY,
    aws_secret_access_key = AWS_SECRET_KEY
)
```

```python
# Create the crawler
create_response = glue.create_crawler(
    Name='stock_data_crawler',
    Role=GLUE_SERVICE_ROLE,
    DatabaseName='stock_data_db',
    Targets={
        'S3Targets': [
            { 'Path': S3_PATH }
        ]
    },
    TablePrefix='stock_'
)

print("Crawler created successfully.")
```

```python
start_response = glue.start_crawler(Name='stock_data_crawler')
print("Crawler started.")
```
